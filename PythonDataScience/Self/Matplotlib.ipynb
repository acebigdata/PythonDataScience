```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,100)
y = np.exp(x)
```

```python
# Functional method
plt.plot(x, y, "r-")
plt.xlabel("X label")
plt.ylabel("Y labe")
plt.title("Title")
```

```python
plt.subplot(2,2,1)
plt.plot(x, y, "r")

plt.subplot(2,2,2)
plt.plot(y, x, "b")

plt.subplot(2,2,3)
plt.plot(x, -y, "g")

plt.subplot(2,2,4)
plt.plot(-y, x, "y")
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x,y)
axes.set_xlabel("X Label")
axes.set_ylabel("Y Label")
axes.set_title ("Title")
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes1.plot(x,y)
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes2.plot(y,x)
```
